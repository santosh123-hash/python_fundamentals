```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
x = [1,2,3]
y = [1,4,9]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.title("Sales indication")
plt.xlabel("Sales per year")
plt.ylabel("revenue per year")
plt.plot(x,y)
plt.show()
```

```python
z = [10,5,0]
```

```python
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
plt.title("Sales indication")
plt.xlabel("Sales per year")
plt.ylabel("revenue per year")
plt.legend(["this is x and y", "this is x and z"])
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

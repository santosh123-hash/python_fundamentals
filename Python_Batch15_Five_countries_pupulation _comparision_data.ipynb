```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data
```

```python
#1.Comparing US and Slovenia countries data
```

```python
US = data[data.country == 'United States']
US
```

```python
Slovenia = data[data.country == 'Slovenia']
Slovenia
```

```python
plt.plot(US.year, US.population)
plt.plot(Slovenia.year, Slovenia.population)
plt.xlabel('year')
plt.ylabel('population')
plt.title('population growth comparion')
plt.legend(['United States', 'Slovenia'])
plt.show()
```
